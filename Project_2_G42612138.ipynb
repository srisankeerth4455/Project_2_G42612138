```python
# importing functions
from heapq import heapify, heappush, heappop
import pandas as pd
import numpy as np
import random
import time
```

```python
# The values of n
n = [10,100,1000,10000,100000]

# creating a list of arrays
list_of_arrays = []
for x in n:
        list_of_arrays.append(sorted((np.random.randint(100,1000, size=(x)))))

print(list_of_arrays)
```

```python
time_taken = []
time_run = []
for i in list_of_arrays:
    start_time = time.time()
    heap_n = []
    heapify(heap_n)
    for j in i:
        heappush(heap_n, j)
    merging_sequence = []
    while len(heap_n) > 1:
        min1 = heappop(heap_n)
        min2 = heappop(heap_n)
        z = min1+min2
        merging_sequence.append(z)
        heappush(heap_n, z)
    end_time = time.time()
    time_taken.append(sum(merging_sequence))
    time_run.append(end_time-start_time)
```

```python
dictionary = {"n":n, "array_count_list":list_of_arrays,"time_taken_for_merge":time_taken, "Time_taken_in_seconds": time_run}
df = pd.DataFrame(dictionary,columns=["n","array_count_list","time_taken_for_merge","Time_taken_in_seconds"])
df
```

```python
# Exporting the DataFrame
df.to_csv('Project 2.csv')
```
